```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df = pd.read_csv("apple_products.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.info()
```

```python
plt.figure(figsize = (15,12))
sns.barplot(df.sort_values(by='Star Rating'),x="Star Rating",y="Product Name")
```

```python
plt.figure(figsize = (15,12))
sns.barplot(df.sort_values(by='Number Of Reviews'),x="Number Of Reviews",y="Product Name")
```

```python
sns.heatmap(df.corr(numeric_only=True),cmap='coolwarm',linewidths=1)
```

```python
#top 10 highest rated iphones
high_rated = df.sort_values(by='Star Rating',ascending=False).head(10)
high_rated['Product Name']
```

```python
#no. of ratings of highest rated iphones
plt.figure(figsize = (15,12))
sns.barplot(high_rated,y='Product Name',x='Number Of Ratings')
```

```python
#no. of reviews of highest rated iphones
plt.figure(figsize = (15,12))
sns.barplot(high_rated,y='Product Name',x='Number Of Reviews')
```

```python
#sale price of highest rated iphones
plt.figure(figsize = (15,12))
sns.barplot(high_rated,y='Product Name',x='Sale Price')
```

```python
plt.figure(figsize=(15,8))
sns.barplot(df,x='Sale Price',y='Number Of Ratings')
```

```python
sns.lmplot(df,x='Sale Price',y='Number Of Ratings')
```

The iphones with lower sale price are sold more

```python
plt.figure(figsize=(15,8))
sns.barplot(df,x='Discount Percentage',y='Number Of Ratings')
```

```python
sns.lmplot(df,x='Discount Percentage',y='Number Of Ratings')
```

Iphone with higher discounts are sold more

Analysis results:
1.APPLE iPhone 8 Plus (Gold, 64 GB) was the most appreciated iPhone in India 
2.iPhones with lower sale prices are sold more in India 
3.iPhones with high discounts are sold more in India
